```python
from sklearn.decomposition import PCA
from utils import kernels, base_indexes, pre_process, feature_extract, train
from sklearn.naive_bayes import GaussianNB
import matplotlib.pyplot as plt
```

```python
models = [
    GaussianNB()
]
```

```python
kernel=kernels[0]
```

```python
def pca_test(indexes: list):
    data_x, data_y = feature_extract(nr_imgs=64, kernel=kernel, indexes=indexes)
    data_x = PCA(n_components=2).fit_transform(data_x)
    train(data_x, data_y, models)
    plt.scatter(data_x[:, 0], data_x[:, 1], c=data_y)
```

```python
plt.figure(figsize=(20, 8))

for i in range(3, 10):
    plt.subplot(2, 4, i-2)
    pca_test(base_indexes[:i])
    
plt.show()
```
